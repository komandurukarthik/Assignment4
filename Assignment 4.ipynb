```python
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
##Connecting to web driver
driver=webdriver.Chrome("chromedriver.exe")
```

```python
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
ranks=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
```

#printing ranks

```python
rank=[]
for i in ranks:
    rank.append(i.text)
```

```python
Names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
```

```python
name=[]
for i in Names:
    name.append(i.text)
```

```python
upload=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
Artist=[]
for i in upload:
    Artist.append(i.text)
```

```python
release=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
upload_date=[]
for i in release:
    upload_date.append(i.text)
```

```python
view=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
views=[]
for i in view:
    views.append(i.text)
```

```python
df=pd.DataFrame()
df['Rank']=rank
df['Video_name']=name
df['Artist']=Artist
df['Upload date']=upload_date
df['views']=views
```

Scrapping team India's international fixtures from bcci.tv

```python
url="https://www.bcci.tv/"
driver.get(url)
```

```python
match_title=driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
title=[]
for i in match_title:
    title.append(i.text)
```

```python
series_name=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
series=[]
for i in series_name:
    series.append(i.text)
```

```python
location=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
```

```python
place=[]
for i in location:
    place.append(i.text.replace("\n"," "))
```

```python
date=[]
day=driver.find_elements_by_xpath("//div[@class='fixture__full-date']")
for i in day:
    date.append(i.text.replace("\n"," "))
```

```python
month=[]
months=driver.find_elements_by_xpath("//span[@class='fixture__month']")
for i in months:
    month.append(i.text.replace("\n"," "))
```

```python
time=[]
clk=driver.find_elements_by_xpath("//span[@class='fixture__time']")
for i in clk:
    time.append(i.text.replace("\n"," "))
```

```python
time
```

```python
Ind=pd.DataFrame()
Ind['Match Title']=title
Ind['Series']=series
Ind['Location']=place
Ind['Date']=date
Ind['match_month']=month
Ind['Time']=time
Ind
```
